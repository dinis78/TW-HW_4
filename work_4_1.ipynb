```python
import numpy as np
from matplotlib import pyplot as plt

a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
a.sort()
plt.show()
```

```python
q_1=np.quantile(a, 0.25)
print(q_1)
```

```python
q_3=np.quantile(a, 0.75)
print(q_3)
```

```python
q_i=q_3-q_1
print(q_i)
```

```python
plt.boxplot(a)
plt.show()
# Выброс всё что больше примерно 100, в нашем случае один выброс 150
```
